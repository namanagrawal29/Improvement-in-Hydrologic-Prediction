```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_excel('Gadarwara_Q.xlsx')
df.drop(columns = ['Unnamed: 38', 'Unnamed: 39', 'Unnamed: 40'], inplace = True)
df
```

```python
part1 = df.drop(df.index[12418:15706])
part2 = df.drop(df.index[0:12418])
print(part1.shape, part2.shape)
```

```python
part1
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
model4 = linear_model.LinearRegression()
model4.fit(part1[['Q(t-1)', 'Q(t-2)', 'RF_5(t-1)', 'RF_6(t-1)', 'RF_8(t-1)']], part1.Q)
Q_pred4 = model4.predict(part2[['Q(t-1)', 'Q(t-2)', 'RF_5(t-1)', 'RF_6(t-1)', 'RF_8(t-1)']])
R2_score4 = r2_score(part2[['Q']], Q_pred4)
MSE4 = mean_squared_error(part2[['Q']], Q_pred4)
RMSE4 = np.sqrt(MSE4)
print(model4.coef_, model4.intercept_, R2_score4, RMSE4)
```
