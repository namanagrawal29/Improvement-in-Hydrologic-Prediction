```python
import pandas as pd
import numpy as np
import math
import statsmodels.api as sm
```

# MLR STARTS HERE

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
# independent variable = RF_1 RF_2 RF_3 RF_4 RF_5 RF_6 RF_8 
model1 = linear_model.LinearRegression()
model1.fit(part1[['RF_1', 'RF_2', 'RF_3', 'RF_4', 'RF_5', 'RF_6', 'RF_8']], part1.Q)
Q_pred1 = model1.predict(part2[['RF_1', 'RF_2', 'RF_3', 'RF_4', 'RF_5', 'RF_6', 'RF_8']])
R2_score1 = r2_score(part2[['Q']], Q_pred1)
MSE1 = mean_squared_error(part2[['Q']], Q_pred1)
RMSE1 = np.sqrt(MSE1)
print(model1.coef_, model1.intercept_, R2_score1, RMSE1)
```
