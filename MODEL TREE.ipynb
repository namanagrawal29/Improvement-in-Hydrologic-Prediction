```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from m5py import M5Prime, export_text_m5
```

```python
df = pd.read_excel('Gadarwara_Q.xlsx')
df
```

```python
part1 = df.drop(df.index[12418:15706])
part2 = df.drop(df.index[0:12418])
print(part1.shape, part2.shape)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
X = part1[['RF_5(t-1)','RF_4(t-1)', 'RF_8(t-1)', 'RF_6(t-1)']].values
Y = part1['Q'].values
regr_1 = M5Prime(use_smoothing=True, use_pruning=True)
regr_2 = M5Prime(use_smoothing=True, use_pruning=False)
regr_3 = M5Prime(use_smoothing=False, use_pruning=True)
regr_4 = M5Prime(use_smoothing=False, use_pruning=False)
regr_1_label = "Tree 1"
regr_2_label = "Tree 2"
regr_3_label = "Tree 3"
regr_4_label = "Tree 4"

regr_1.fit(X, Y)
regr_2.fit(X, Y)
regr_3.fit(X, Y)
regr_4.fit(X, Y)

X_test = part2[['RF_5(t-1)','RF_4(t-1)', 'RF_8(t-1)', 'RF_6(t-1)']].values
Y_test = part2['Q'].values
Y_pred1 = regr_1.predict(X_test)
Y_pred2 = regr_2.predict(X_test)
Y_pred3 = regr_3.predict(X_test)
Y_pred4 = regr_4.predict(X_test)

R2_score1 = r2_score(Y_test, Y_pred1)
R2_score2 = r2_score(Y_test, Y_pred2)
R2_score3 = r2_score(Y_test, Y_pred3)
R2_score4 = r2_score(Y_test, Y_pred4)

print("R2_score1 = ", R2_score1)
print("R2_score2 = ", R2_score2)
print("R2_score3 = ", R2_score3)
print("R2_score4 = ", R2_score4)
```
